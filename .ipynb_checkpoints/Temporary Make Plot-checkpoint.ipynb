```python
import pandas as pd 
import numpy as np
```

```python
interested_df = pd.read_csv('interested_df.csv')
stock_data = pd.read_csv('stock_data.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
interested_df['stock_ticker'].unique()
```

```python
fig, ax = plt.subplots(2,interested_df['stock_ticker'].nunique(),sharex=True)
```

```python
stocks = interested_df['stock_ticker'].unique()

fig, ax = plt.subplots(2,interested_df['stock_ticker'].nunique(),sharex='col')

for i in range(len(stocks)):
    data_n = stock_data[stock_data['symbol'] == stocks[i]]
    ax[0,i].plot(data_n['date'], data_n['adj_close'])
    ax[1,i].plot(data_n['date'], data_n['adj_volume'])
    
    ax[0, i].title.set_text(stocks[i])
    
    
plt.tight_layout()
```

```python
first_row = stock_data.groupby(['symbol']).first()[['date', 'adj_close']]

first_row
```

```python
initial_data = interested_df[['stock_ticker', 'value_invested', 'start_date', 'end_date']].merge(first_row, left_on='stock_ticker', right_on=first_row.index)

initial_data['unit_bought'] = initial_data['value_invested'] / initial_data['adj_close']

initial_data = initial_data[['stock_ticker', 'date', 'end_date', 'unit_bought']]

initial_data.rename(columns={'date': 'start_date'}, inplace=True)

initial_data
```

```python
combined_data = stock_data[['symbol', 'adj_close', 'date']].merge(initial_data, left_on='symbol', right_on='stock_ticker')

combined_data
```

```python
combined_data['total_value'] = combined_data['adj_close'] * combined_data['unit_bought']

combined_data
```

```python
combined_data.groupby(['date'])['total_value'].sum().to_frame()
```
